```python
import pandas as pd

# Path to your file in Google Drive
file_path = 'UNSW_courses.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
def concatenate_details(row):
    details = []
    if pd.notna(row['Course Description']) and row['Course Description'] != '':
        details.append('Course Description: ' + row['Course Description'])
    if pd.notna(row['Course Aims']) and row['Course Aims'] != '':
        details.append('Course Aims: ' + row['Course Aims'])
    if pd.notna(row['Course Learning Outcomes']) and row['Course Learning Outcomes'] != '':
        details.append('Course Learning Outcomes: ' + row['Course Learning Outcomes'])
    return '\n'.join(details)

# Apply the function across the DataFrame rows
df['details'] = df.apply(concatenate_details, axis=1)

#df.to_csv('full_courses.csv', index=False)

# Display the modified DataFrame
#print(df.head(5))
```

```python
condition = (
    ~df['Academic Unit'].str.contains('School of Computer Science and Engineering') &
    ~df['Academic Unit'].str.contains('School of Mathematics & Statistics') &
    ~df['Academic Unit'].str.contains('School of Engineering and Technology')
)

# Apply the condition to filter the DataFrame
filtered_df = df[condition]
#print(filtered_df.iloc[1655])
```

```python
async def get_responses(api_key, messages):
    responses = []
    async for partial in fp.get_bot_response(messages=messages, bot_name="GPT-3.5-Turbo", api_key=api_key):
        # Parse the response to extract the sentiment text
        try:
            # Assuming 'raw_response' is a dictionary containing 'text' as a JSON string
            response_content = json.loads(partial.raw_response['text'])
            responses.append(response_content['text'])
            full_response = ''.join(responses)
        except (KeyError, json.JSONDecodeError):
            # Handle cases where parsing fails or the expected keys are not present
            responses.append('Error parsing response')
    return full_response

# Replace <api_key> with your actual API key, ensuring it is a string.
api_key = 'W7Yys-qmxRHpc7pos-j7Fv4sQcNg4yAkqwF-ejCgN6o'

skills_list = []

for detail in filtered_df['details'].iloc[:5]:
    text = detail
    message = fp.ProtocolMessage(role="user", content=f"Please extract three key skills from {text}. Each key skills should be strictly less than 7 words and please make the response strictly in the format of: Skills: skill1; skill2; skill3. Text only and no listing numbers.")
    # Run the event loop
    async def main():
        response = await get_responses(api_key, [message])
        skills_list.append(response)
        print(response)

    # Execute the main function asynchronously
    await main()

filtered_df.loc[:, 'Skills'] = skills_list

# Run the event loop
#async def main():
    #response = await get_responses(api_key, [message])
    #print(response)

# Execute the main function asynchronously
#await main()
```
